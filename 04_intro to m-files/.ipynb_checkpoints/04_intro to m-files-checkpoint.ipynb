## Functions

So far, everything has been executed as a script, or calling a built-in function. 

Now we begin building our own functions.

Functions are saved in memory (or better yet) in a folder in your path or current directory

Example of storing a function in memory

$f(x,y) = (xy^{3}-x^{3}y)$

````octave
f= @(x,y) (x.*y.^3-x.^3.*y)
````

````octave
x=linspace(-1,1,10); % 10 data points from -1 to 1
y=linspace(-1,1,10); % 10 data points from -1 to 1
[X,Y]=meshgrid(x,y); % 100 data points from -1 to 1 on the x- and y-axes
Z=f(X,Y);            % Z=f(X,Y) evaluated at the 100 data points

mesh(X,Y,Z)
````

Here we will save a function called `my_velocity` as `my_velocity.m`

```matlab 
function [vx,vy] = my_velocity(x,y,t)
    % Help documentation of "my_velocity"
    % This function computes the velocity in the x- and y-directions given
    % three vectors of position in x- and y-directions as a function of time
    % x = x-position
    % y = y-position
    % t = time
    % output
    % vx = velocity in x-direction
    % vy = velocity in y-direction
    
    vx=zeros(length(t),1);
    vy=zeros(length(t),1);
    
    vx(1:end-1) = diff(x)./diff(t); % calculate vx as delta x/delta t
    vy(1:end-1) = diff(y)./diff(t); % calculate vy as delta y/delta t
    
    vx(end) = vx(end-1);
    vy(end) = vy(end-1);

end
```

````octave
help my_velocity
````

````octave
t=linspace(0,10,100)'; 
x=t.^3; % vx = 3*t^2
y=t.^2/2;  % vy = t
[vx,vy]=my_velocity(x,y,t);
````

````octave
[axes,h1,h2] = plotyy(t(1:10:end),vx(1:10:end), t(1:10:end),vy(1:10:end));
set([h1,h2],"marker","o")
set([h1,h2],"linewidth",0)
hold on
[~,h3,h4]=plotyy(t,3*t.^2,t,1*t);
xlabel ("time");
ylabel (axes(1), "v_{x}");
ylabel (axes(2), "v_{y}");
````

Create a new function that calls 'my_velocity' called, `my_acceleration.m`

```matlab
function [ax,ay]=my_acceleration(x,y,t)
    % Help documentation of "my_acceleration"
    % This function computes the acceleration in the x- and y-directions given
    % three vectors of position in x- and y-directions as a function of time
    % x = x-position
    % y = y-position
    % t = time
    % output
    % ax = acceleration in x-direction
    % ay = acceleration in y-direction

    function v=diff_match_dims(x,t)
      v=zeros(length(t),1);
      v(1:end-1)=diff(x)./diff(t);
      v(end)=v(end-1);
    end

    [vx,vy]=my_velocity(x,y,t);

    ax = diff_match_dims(vx,t);
    ay = diff_match_dims(vy,t);

end
```

````octave
help my_acceleration
````

````octave
[ax,ay]=my_acceleration(x,y,t);
````

````octave
[axes,h1,h2] = plotyy(t(1:10:end),ax(1:10:end), t(1:10:end),ay(1:10:end));
set([h1,h2],"linewidth",0)
set([h1,h2],"marker",'s')
hold on
[~,h3,h4]=plotyy(t,6*t,t,1*t./t);
xlabel ("time");
ylabel (axes(1), "a_{x}");
ylabel (axes(2), "a_{y}");
````
