# Splines



Following interpolation discussion, instead of estimating 9 data points with an eighth-order polynomial, it makes more sense to fit sections of the curve to lower-order polynomials:

0. zeroth-order (nearest neighbor)

1. first-order (linear interpolation)

2. third-order (cubic interpolation)

Matlab and Octave have built-in functions for 1D and 2D interpolation:

`interp1`

`interp2`

```octave
help interp1
```

# Choosing 

```octave
x=linspace(-pi,pi,9);
xi=linspace(-pi,pi,100);
y=sin(x);
yi_lin=interp1(x,y,xi,'linear');
yi_spline=interp1(x,y,xi,'spline'); 
yi_cubic=interp1(x,y,xi,'cubic');
plot(x,y,'o',xi,yi_lin,xi,yi_spline,xi,yi_cubic)
axis([-pi,pi,-1.5,1.5])
legend('data','linear','cubic spline','piecewise cubic','Location','NorthWest')
```

### Example: Accelerate then hold velocity

Test driving a car, the accelerator is pressed, then released, then pressed again for 20-second intervals, until speed is 120 mph. Here the time is given as vector t in seconds and the velocity is in mph. 

```octave
t=[0 20 40 68 80 84 96 104 110]';
v=[0 20 20 80 80 100 100 125 125]';
tt=linspace(0,110)';
v_lin=interp1(t,v,tt);
v_spl=interp1(t,v,tt,'spline');
v_cub=interp1(t,v,tt,'cubic');

plot(t,v,'o',56,38,'s',tt,v_lin,tt,v_spl,tt,v_cub)
xlabel('t (s)')
ylabel('v (mph)')
legend('data','removed data point','linear','cubic spline','piecewise cubic','Location','NorthWest')
```

```octave
t=[0 20 40 56 68 80 84 96 104 110]';
v=[0 20 20 38 80 80 100 100 125 125]';
tt=linspace(0,110)';
v_lin=interp1(t,v,tt);
v_spl=interp1(t,v,tt,'spline');
v_cub=interp1(t,v,tt,'cubic');


plot(tt(2:end),diff(v_lin)./diff(tt),tt(2:end),diff(v_spl)./diff(tt),tt(2:end),diff(v_cub)./diff(tt))
xlabel('t (s)')
ylabel('dv/dt (mph/s)')
legend('linear','cubic spline','piecewise cubic','Location','NorthWest')
```

### Choose spline wisely

For example of sin(x), not very important

For stop-and-hold examples, the $C^{2}$-continuity should not be preserved. You don't need smooth curves.



# Numerical Integration

A definite integral is defined by 

$I=\int_{a}^{b}f(x)dx$

To determine the mass of an object with varying density, you can perform a summation

mass=$\sum_{i=1}^{n}\rho_{i}\Delta V_{i}$

or taking the limit as $\Delta V \rightarrow dV=dxdydz$

mass=$\int_{0}^{h}\int_{0}^{w}\int_{0}^{l}\rho(x,y,z)dxdydz$

## Newton-Cotes Formulas

$I=\int_{a}^{b}f(x)dx=\int_{a}^{b}f_{n}(x)dx$

where $f_{n}$ is an n$^{th}$-order polynomial approximation of f(x)

## First-Order: Trapezoidal Rule

$I=\int_{a}^{b}f(x)dx\approx \int_{a}^{b}\left(f(a)+\frac{f(b)-f(a)}{b-a}(x-a)\right)dx$

$I\approx(b-a)\frac{f(a)+f(b)}{2}$

```octave
x=linspace(0,pi)';
plot(x,sin(x),[0,pi/2],sin([0,pi/2]))
I_trap=mean(sin(([0,pi/2]))*(diff([0,pi/2])))
I_act = -(cos(pi/2)-cos(0))
```

Improve estimate with more points

$I=\int_{a}^{b}f(x)dx=\int_{a}^{a+\Delta x}f(x)dx+\int_{a+\Delta x}^{a+2\Delta x}f(x)dx+ \cdots \int_{b-\Delta x}^{b}f(x)dx$

$I\approx\Delta x\frac{f(a)+f(a+\Delta x)}{2}+\Delta x\frac{f(a+\Delta x)+f(a+2\Delta x)}{2}
+\cdots \Delta x\frac{f(b-\Delta x)+f(b)}{2}$

$I\approx \frac{\Delta x}{2}\left(f(a)+2\sum_{i=1}^{n-1}f(a+i\Delta x) +f(b)\right)$

```octave
N=70;
I_trap=trap(@(x) sin(x),0,pi/2,N);
fprintf('For %i steps\ntrapezoid approximation of integral is %1.2f \n actual integral is %1.2f',N,I_trap,I_act)
```

```octave
x=linspace(0,pi);
plot(x,sin(x),linspace(0,pi/2,N),sin(linspace(0,pi/2,N)),'-o')
```

## Increase accuracy = Increase polynomial order

### Simpson's Rules

When integrating f(x) and using a second order polynomial, this is known as **Simpson's 1/3 Rule**

$I=\frac{h}{3}(f(x_{0})+4f(x_{1})+f(x_{2}))$

where a=$x_{0}$, b=$x_{2}$, and $x_{1}=\frac{a+b}{2}$

This can be used with n=3 or multiples of 2 intervals

$I=\int_{x_{0}}^{x_{2}}f(x)dx+\int_{x_{2}}^{x_{4}}f(x)dx+\cdots +\int_{x_{n-2}}^{x_{n}}f(x)dx$

$I=(b-a)\frac{f(x_{0})+4\sum_{i=1,3,5}^{n-1}f(x_{i})+2\sum_{i=2,4,6}^{n-2}f(x_{i})+f(x_{n})}{3n}$

```octave
f=@(x) 0.2+25*x-200*x.^2+675*x.^3-900*x.^4+400*x.^5;
simpson3(f,0,0.8,4)
Is_1_3=simpson3(@(x) sin(x),0,pi/2,2)
```
